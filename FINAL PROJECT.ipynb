```python
df = df.dropna()
df.shape
```

```python
Q1 = df[numerik].quantile(0.25)
Q3 = df[numerik].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

#Filter baris outliner
df_outlier = df[((df[numerik] < lower_bound) | (df[numerik] > upper_bound)).any(axis=1)]
print('Data outlier: ', df_outlier.shape[0], 'baris')
```

# Feature Engineering

```python
# Purchase Frequency
df['customer_frequency'] = df.groupby('user_id')['user_id'].transform('count')
df.head()
```

```python
# Age group segmentation
df['age_group'] = pd.cut(
    df['age'],
    bins=[0, 17, 25, 35, 50, 100],
    labels=['<18', '18-25', '26-35', '36-50', '50+']
)
df.head()
```

```python
# AVG order value
df['avg_order_value'] = df.groupby('user_id')['total_purchase'].transform('mean')
df.head()
```

```python
# Loyalty Tier Classification
df['loyalty_segment'] = df['loyalty_tier'].apply(
    lambda x: (
        'No Loyalty' if pd.isna(x) else
        'Bronze' if x == 1 else
        'Silver' if x == 2 else
        'Gold' if x == 3 else
        'Platinum'
    )
)
df.head()
```
